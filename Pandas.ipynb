**Pandas**

Pandas is a Python library used to analyze data.
The name "Pandas" has a reference to both "Panel Data", and "Python Data Analysis" and was created by Wes McKinney in 2008.

```python
import pandas as pd
print(pd.__version__)
```

> Series

It is for 1D array

```python
cars = ['Toyota','Suzuki','BMW']
print(cars)
cars = pd.Series(cars)
print(cars)
```

> Dataframe

It is for 2D array

```python
cars_dt = {
    'name':['Toyota','Suzuki','BMW'],
    'model':[2022,2019,2019]
}
print(cars_dt)
print("______")
print(pd.DataFrame(cars_dt))
```
